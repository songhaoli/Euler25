Euler P25. What is the index of the first term in the Fibonacci sequence to contain 1000 digits?

```
import math
```

```
# The golden ratio
phi = (5**(1/2)+1) / 2

# Returns the n'th term in Fibonacci sequence
def fib(n):
  f = phi**n/5**(1/2) - (1-phi)**n/5**(1/2)
  return round(f)
```

```
fib(10)
```

```
# Split fib(n) into two pieces, fib1(n) and fib(2)

# fib1(n) converges to fib(n)
def fib1(n):
  f1 = phi**n/5**(1/2)
  return round(f1,4)

# fib2(n) converges to 0
def fib2(n):
  f2 = (1-phi)**n/5**(1/2)
  return round(f2,4)
```

```
fib1(20)
```

```
fib2(20)
```

So for very large n, only fib1 matters.

In other words, we need to solve the equation:

\left( \frac{\phi}{\sqrt{5}} \right)^n = 10^999

That is,

n = \frac{999 \ln 10}{\ln \frac{\phi}{\sqrt{5}}}


```
n = (999*math.log(10)+math.log(5**(1/2)))/math.log(phi)
print(n)
print(int(n)+1)
```

*Answer*. 4782
